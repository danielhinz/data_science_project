##### 2) Wilhelmstraße

```python
url = 'https://www.my-stuwe.de//wp-json/mealplans/v1/canteens/715?lang=de&v=1729619513869'

# Make the GET request
response = requests.get(url)

# Parse the JSON response
json_data = response.json()

# Pretty print the JSON response
pretty_json = json.dumps(json_data, indent=4)
print(pretty_json)

# Create and print df
df = pd.json_normalize(json_data)

# Extract the list of dictionaries from the column
menus_list = df['715.menus'].iloc[0]

# Convert the list of dictionaries into a new DataFrame
menus_df = pd.DataFrame(menus_list)

# Save to excel
menus_df.to_excel('menus_cafeteria_715.xlsx', index=False)

# Get cuda
device = "cuda" if torch.cuda.is_available() else "cpu"

# Create the 'images_prinzkarl' folder if it doesn't exist
if not os.path.exists('images_cafeteria_715'):
    os.makedirs('images_cafeteria_715')

# Iterate over each row in the DataFrame
for index, row in menus_df.iterrows():
    # Set up the prompt
    prompt = row["menu"]
    
    try:
        # Generate the image
        image = pipeline(prompt=prompt).images[0]
        
        # Define the filename using the 'id'
        filename = f"{row['id']}_image.png"
        
        # Create the full path for saving the image
        filepath = os.path.join('images_cafeteria_715', filename)
        
        # Save the generated image
        image.save(filepath)
        
        # Append the filename to a new column in the DataFrame
        menus_df.at[index, 'image_filename'] = filename
        print(f"Image saved: {filepath}")
    except Exception as e:
        print(f"Error processing row {index}: {str(e)}")
        menus_df.at[index, 'image_filename'] = "error"

# Display the updated DataFrame
print(menus_df)
```

##### 4) Cafeteria Unibibiothek

```python
URL = "https://www.my-stuwe.de/mensa/cafeteria-unibibliothek-tuebingen/"
page = requests.get(URL).content # get page
tree = etree.HTML(page) # store hierarchy

# Find all <p> elements
p_elements = tree.xpath("//div[@class='avia_textblock']")

# Extract text content from <p> elements and their descendants
extracted_content = []
for p in p_elements:
    # Get text content of the <p> element and all its descendants
    content = ''.join(p.itertext()).strip()
    if content:  # Only add non-empty content
        extracted_content.append(content)

# Print or process the extracted content
for item in extracted_content:
    print(item)
```
